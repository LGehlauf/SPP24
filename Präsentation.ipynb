## Fertigungsaufbau
- Werkstattfertigung mit 5 Betriebsmittelgruppen (jeweils 1-2 Betriebsmittel)
- 7 Produkte mit ihren jeweiligen Arbeitsplänen in DB definiert 
- Erzeugung der Aufträge über Bestellbestandsverfahren entsprechend der generierten Entnahmen aus Fertigteillager 
- Auftragsfreigabe durch Engpasssteuerung implementiert
- Transporte durch drei Transportfahrzeuge (jeweils mit unteschiedlichen Geschwindigkeiten und Akkukapazitäten) ausgeführt 
- Definition eines Graphen und Einsatz des Dijkstra-Algorithmus zur Berechnung der kürzesten Transportwege
- Simulation von Auschüssen durch Ziehen von Zufallszahlen aus Binomialverteilung
- Simulation von Bearbeitungszeiten durch Ziehen von Zufallszahlen aus Log.-Normalverteilung
- Simulation von Maschinenausfällen durch Ziehen von Zufallszahlen aus Weibullverteilung 
- Implementierung eines Betriebskalenders, welcher Feiertage und Wochenende sowie wählbare Schichtregime berücksichtigt und die Produktion dementsprechend pausiert 
- Möglichkeit zur Defnition einer Liste zusätzlicher Aufträge
